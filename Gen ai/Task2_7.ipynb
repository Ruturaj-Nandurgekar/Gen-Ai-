```python
# Task 2
!pip install emoji
import re
import emoji

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return ""

    # 1. Convert to lowercase
    text = text.lower()

    # 2. Remove URLs and Emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S+@\S+', '', text)

    # 3. Handle Emojis
    text = emoji.replace_emoji(text, replace='')

    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)

    # 5. Handle repeated characters
    # This regex looks for characters repeated 3 or more times and reduces to 1
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 6. Remove punctuation and special symbols
    text = re.sub(r'[^\w\s]', '', text)

    # 7. Remove extra spaces
    text = " ".join(text.split())

    # 8. Tokenize and filter by length
    tokens = text.split()
    keep_words = {'no', 'not'}
    cleaned_tokens = [t for t in tokens if len(t) > 2 or t in keep_words]

    return cleaned_tokens
```

```python
#  Task 3
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",
    "",                # Edge Case: Empty
    "😊😇🔥",          # Edge Case: Only Emojis
    "12345 678"        # Edge Case: Only Numbers
]

print(f"{'Original Text':<40} | {'Cleaned Sentence'}")
print("-" * 80)

processed_data = []
for item in sample_inputs:
    tokens = preprocess_text(item)
    clean_sent = " ".join(tokens)
    processed_data.append({"original": item, "tokens": tokens, "clean": clean_sent})
    print(f"{item[:38]:<40} | {clean_sent}")
```

```python
# Task 4
from collections import Counter

all_all_tokens = []

print(f"\n{'Cleaned Sentence':<30} | {'Total'} | {'Unique'} | {'Avg Len'}")
print("-" * 70)

for entry in processed_data:
    tks = entry['tokens']
    if not tks:
        continue

    total = len(tks)
    unique = len(set(tks))
    avg_len = sum(len(t) for t in tks) / total
    all_all_tokens.extend(tks)

    print(f"{entry['clean'][:30]:<30} | {total:<5} | {unique:<6} | {avg_len:.2f}")
```

```python
# Task 5
counts = Counter(all_all_tokens)
print("\nTop 10 Most Frequent Words:", counts.most_common(10))
print("Top 5 Least Frequent Words:", counts.most_common()[:-6:-1])
```

```python
# Task 6 & 7
def full_pipeline(text_list):
    """
    Standardizes the processing of a list of strings.
    Handles empty lists and non-string inputs gracefully.
    """
    if not text_list:
        return {"tokens": [], "clean_sentences": []}

    pipeline_result = {
        "tokens": [],
        "clean_sentences": []
    }

    for text in text_list:
        # Error handling for non-string types
        if not isinstance(text, str):
            continue

        tokens = preprocess_text(text)


        if tokens:
            pipeline_result["tokens"].append(tokens)
            pipeline_result["clean_sentences"].append(" ".join(tokens))

    return pipeline_result

# Execution
final_output = full_pipeline(sample_inputs)
print("\nFinal Pipeline Tokens (Snippet):", final_output["tokens"][:3])
```
